 Q1. Explain the different types of data (qualitative and quantitative) and provide examples of each. Discuss
nominal, ordinal, interval, and ratio scales.
- Types of Data: Qualitative and Quantitative
1. Qualitative Data: This type of data describes characteristics or qualities and is often non-numeric. Examples include:

- Nominal Scale: Categories without a specific order (e.g., colors like red, blue, green; types of fruit like apple, banana, orange).
- Ordinal Scale: Categories with a meaningful order but no consistent difference between them (e.g., rankings like 1st, 2nd, 3rd; satisfaction ratings like poor, fair, good, excellent).
2. Quantitative Data: This type of data is numeric and can be measured. Examples include:

- Interval Scale: Numeric data with meaningful intervals but no true zero point (e.g., temperature in Celsius or Fahrenheit).
- Ratio Scale: Numeric data with a true zero point, allowing for the comparison of magnitudes (e.g., height, weight, age, income).



 Q2. What are the measures of central tendency, and when should you use each? Discuss the mean, median,
and mode with examples and situations where each is appropriate.
 - Measures of Central Tendency
1. Mean: The average of a dataset. Use when data is normally distributed without outliers. Example: Average test score of a class.
2. Median: The middle value when data is ordered. Use when data is skewed or has outliers. Example: Median income in a region.
3. Mode: The most frequently occurring value. Use when data is categorical or to identify the most common item. Example: Most common shoe size sold in a store.





 Q3. Explain the concept of dispersion. How do variance and standard deviation measure the spread of data?
 - Dispersion refers to the spread of data points in a dataset.

1. Variance: Measures the average squared deviation from the mean. A higher variance indicates more spread.
2. Standard Deviation: The square root of variance, providing a measure of spread in the same units as the data. A higher standard deviation indicates more variability.



 Q4. What is a box plot, and what can it tell you about the distribution of data?
 - A box plot (or whisker plot) visually represents the distribution of data based on five summary statistics: minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum. It can show:

1. The central tendency (median).
2. The spread (IQR).
3. Potential outliers (points outside 1.5 times the IQR).



 Q5. Discuss the role of random sampling in making inferences about populations.
 - Random sampling is crucial for making inferences about populations because it ensures that every individual has an equal chance of being selected. This reduces bias and allows for generalizations from the sample to the population, enhancing the validity of statistical conclusions.



 Q6. Explain the concept of skewness and its types. How does skewness affect the interpretation of data?
 - Skewness measures the asymmetry of the distribution of data:

1. Positive Skew: Tail on the right side; mean > median.
2. Negative Skew: Tail on the left side; mean < median. Skewness affects interpretation by indicating whether the mean is a good representation of the data. In skewed distributions, the median may be a better measure of central tendency.




 Q7. What is the interquartile range (IQR), and how is it used to detect outliers?
 - The IQR is the range between the first quartile (Q1) and the third quartile (Q3) and is used to measure the spread of the middle 50% of data. It helps detect outliers by identifying values that fall below Q1 - 1.5 * IQR or above Q3 + 1.5 * IQR.





 Q8. Discuss the conditions under which the binomial distribution is used.
 - The binomial distribution is used when:

1. There are a fixed number of trials (n).
2. Each trial has two possible outcomes (success or failure).
3. The probability of success (p) is constant across trials.
4. Trials are independent.





 Q9. Explain the properties of the normal distribution and the empirical rule (68-95-99.7 rule).
 - The normal distribution is symmetric, bell-shaped, and characterized by its mean (μ) and standard deviation (σ). The empirical rule states:

1. Approximately 68% of data falls within one standard deviation of the mean.
2. About 95% falls within two standard deviations.
3. About 99.7% falls within three standard deviations

 Q11. Explain what a random variable is and differentiate between discrete and continuous random variables.
 - Definition: A random variable is a variable whose value is determined by the outcome of a random process. It assigns numerical values to the outcomes of an experiment.

- Types of Random Variables:

1. Discrete Random Variables:
- Take on a countable number of distinct values.
- Examples include the number of heads in a series of coin tosses or the result of rolling a die.
2. Continuous Random Variables:
- Can take any value within a given range or interval.
- Examples include measurements like height, weight, or temperature, where values can be infinitely precise.






 Q12. Provide an example dataset, calculate both covariance and correlation, and interpret the results
 - Example Dataset:

X                       Y


1                       2


2                       3

4                       4

- Calculating Covariance:

- Formula: [ \text{Cov}(X, Y) = \frac{1}{n-1} \sum (X_i - \bar{X})(Y_i - \bar{Y}) ]
- Steps:
1. Calculate the means (\bar{X}) and (\bar{Y}).
2. Compute the products of the deviations from the mean for each pair.
3. Sum these products and divide by (n-1).
- Calculating Correlation:

- Formula: [ r = \frac{\text{Cov}(X, Y)}{\sigma_X \sigma_Y} ]
- Where:
(\sigma_X) and (\sigma_Y) are the standard deviations of X and Y, respectively.
- Interpretation of Results:

1. A positive covariance indicates that as X increases, Y tends to increase as well.
2. A correlation coefficient close to +1 suggests a strong positive linear relationship, meaning that the values of X and Y move together in the same direction.
